## 학습목표
 1. DataFrame groupby 이해하기

```python
import pandas as pd
import numpy as np
```

#### group by
  + 아래의 세 단계를 적용하여 데이터를 그룹화(groupping) (SQL의 group by 와 개념적으로는 동일, 사용법은 유사)
    - 데이터 분할
    - operation 적용
    - 데이터 병합

```python
# data 출처: https://www.kaggle.com/hesh97/titanicdataset-traincsv/data
df = pd.read_csv('./data/train.csv')
df.head()
```

```python
df[df['Age']<1]
```

#### GroupBy groups 속성
 - 각 그룹과 그룹에 속한 index를 dict 형태로 표현

```python
class_group=df.groupby('Pclass')
```

```python
class_group
```

```python
class_group.groups
```

```python
class_group.mean()['Survived']
```

```python
gender_group=df.groupby('Age')
gender_group.groups
```

#### groupping 함수
 - 그룹 데이터에 적용 가능한 통계 함수(NaN은 제외하여 연산)
 - count - 데이터 개수 
 - sum   - 데이터의 합
 - mean, std, var - 평균, 표준편차, 분산
 - min, max - 최소, 최대값

```python
gender_group.mean()['Pclass']
```

```python
gender_group.mean()['Survived']
```

```python
gender_group.mean().index
```

* 성별에 따른 생존율 구해보기

#### 복수 columns로 groupping 하기
 - groupby에 column 리스트를 전달
 - 통계함수를 적용한 결과는 multiindex를 갖는 dataframe

* 클래스와 성별에 따른 생존률 구해보기

```python
df.groupby(['Pclass','Sex']).mean()
```

```python
df.groupby(['Pclass','Sex']).mean().index
```

#### index를 이용한 group by
 - index가 있는 경우, groupby 함수에 level 사용 가능
   - level은 index의 depth를 의미하며, 가장 왼쪽부터 0부터 증가

* **set_index** 함수
 - column 데이터를 index 레벨로 변경
* **reset_index** 함수
 - 인덱스 초기화

```python
df.set_index(['Pclass','Sex'])
```

```python
df.head()
```

```python
df.set_index('Age')
```

```python
df.groupby('Age').mean()
```

```python
df.set_index('Age').groupby(level=0).mean()  # => level=0은 index 0번째를 기준으로 그룹을 지어라
```

### 나이대별로 생존율 구하기


```python
import math
def age_categorize(age):
    if math.isnan(age):
        return -1
    return math.floor(age/10)*10
```

```python
df.set_index('Age').groupby(age_categorize).mean()
```

```python
df.set_index('Age').groupby(age_categorize).mean()['Survived']  # index가 있는 경우엥 그룹을 지을 수 있다.
```

#### MultiIndex를 이용한 groupping

```python
df.set_index(['Pclass','Sex']).groupby(level=[0,1]).mean()['Age']
```

#### aggregate(집계) 함수 사용하기
 - groupby 결과에 집계함수를 적용하여 그룹별 데이터 확인 가능

```python
df.set_index(['Pclass','Sex']).groupby(level=[0,1]).aggregate([np.sum,np.mean])
```

```python
df.groupby('Age').aggregate([np.sum,np.mean])
```

### 굳이 set_index를 해야되는지?

```python
exam_data = {'이름' : ['서준','우현','인아'],
            '수학' : [90,80,70],
            '영어' : [98,89,95],
            '음악' : [85,95,100],
            '체육' : [100,90,90]}
df = pd.DataFrame(exam_data)
```

```python
df
```

```python
df.groupby('수학').agg([np.max,np.mean])
```

```python
df.groupby('이름').agg([np.max,np.mean])
```

```python
df.set_index('이름').groupby(level=0).aggregate([np.max,np.mean])
```
